```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score, KFold
from sklearn.impute import SimpleImputer
from sklearn.metrics import accuracy_score

# Load training data
train_data = pd.read_csv('Training.csv')
train_data.drop('Unnamed: 133', axis=1, inplace=True)
X_train = train_data.drop('prognosis', axis=1)
y_train = train_data['prognosis']

# Load testing data
test_data = pd.read_csv('Testing.csv')
X_test = test_data.drop('prognosis', axis=1)
y_test = test_data['prognosis']

# Handle missing values
imputer = SimpleImputer(strategy='mean')  # You can choose a different strategy if needed
X_train_imputed = imputer.fit_transform(X_train)
X_test_imputed = imputer.transform(X_test)

# Initialize the Classifier
clf = RandomForestClassifier(random_state=42)

# Train the model
clf.fit(X_train_imputed, y_train)

# Get feature importances
feature_importances = clf.feature_importances_

# Create a DataFrame to show feature importances
feature_importance_df = pd.DataFrame({'Feature': X_train.columns, 'Importance': feature_importances})

# Sort the DataFrame by importance
sorted_feature_importance = feature_importance_df.sort_values(by='Importance', ascending=False)

# Print the top features
print(sorted_feature_importance.head(40))  # Adjust the number as
```

```python
import matplotlib.pyplot as plt

# Assuming selected_features is a list of feature names
# and X_train is your training data with these features

# Define the number of columns and rows for the subplots
num_features = len(sorted_feature_importance)
num_cols = 3  # Adjust the number of columns as needed
num_rows = (num_features + num_cols - 1) // num_cols

# Create subplots
fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 3*num_rows))

# Flatten the axes array for easier iteration
axes = axes.flatten()

# Loop over selected features and create subplots
for i, feature in enumerate(selected_features):
    ax = axes[i]
    ax.hist(X_train[feature], bins=20)  # Assuming you want to create histograms
    ax.set_title(f'Distribution of {feature}')
    ax.set_xlabel(feature)
    ax.set_ylabel('Frequency')

# Hide any extra subplots if needed
if len(selected_features) < num_rows * num_cols:
    for j in range(len(selected_features), num_rows*num_cols):
        fig.delaxes(axes[j])

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```

```python
# Assuming 'selected_features' contains the list of selected columns
# selected_features = ['itching', 'skin_rash', 'joint_pain', 'vomiting', 'fatigue',
#                      'high_fever', 'headache', 'loss_of_appetite', 'chest_pain', 'irritability']
selected_features = sorted_feature_importance['Feature'][:40]
# Create the feature set 'x' using only the selected columns
x = df[selected_features]

# Create the target variable 'y'
y = df['prognosis']

# Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.5, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# Initialize the Decision Tree Classifier
dt_classifier = DecisionTreeClassifier(random_state=42)

# Train the model
dt_classifier.fit(x_train[selected_features], y_train)

# Predict using the same features
y_pred_train = dt_classifier.predict(x_train[selected_features])

# Calculate accuracy on training data
train_accuracy = accuracy_score(y_train, y_pred_train)
print(f'Accuracy on Training Data: {train_accuracy:.2f}')

# Predict on test data
y_pred_test = dt_classifier.predict(x_test[selected_features])

# Calculate accuracy on test data
test_accuracy = accuracy_score(y_test, y_pred_test)
print(f'Accuracy on Test Data: {test_accuracy:.2f}')
```
